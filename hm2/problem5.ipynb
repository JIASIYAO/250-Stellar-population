```python
%matplotlib inline
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pdb
import fsps
```

```python
sp1 = fsps.StellarPopulation(imf_type=2, imf3=0.8, sf_start=0.01)
wave1, spec1=sp1.get_spectrum(tage = 0.01)
sp2 = fsps.StellarPopulation(imf_type=2, imf3=1.3, sf_start=0.01)
wave2, spec2=sp2.get_spectrum(tage = 0.01)
sp3 = fsps.StellarPopulation(imf_type=2, imf3=1.8, sf_start=0.01)
wave3, spec3=sp3.get_spectrum(tage = 0.01)
idx = np.where((wave1>1500) & (wave1<10000))[0]
plt.clf()
plt.plot(wave1[idx], spec1[idx], label='alpha = 0.8')
plt.plot(wave2[idx], spec2[idx], label='alpha = 1.3')
plt.plot(wave3[idx], spec3[idx], label='alpha = 1.8')
plt.legend()
plt.xlabel('wavelength')
plt.ylabel('flux')
plt.title('spectrum')
```

```python
sp1 = fsps.StellarPopulation(imf_type=0, sf_start=10)
wave1, spec1=sp1.get_spectrum(tage = 10)
sp2 = fsps.StellarPopulation(imf_type=2, sf_start=10)
wave2, spec2=sp2.get_spectrum(tage = 10)
sp3 = fsps.StellarPopulation(imf_type=3, sf_start=10)
wave3, spec3=sp3.get_spectrum(tage = 10)
idx = np.where((wave1>5000) & (wave1<20000))[0]
plt.clf()
plt.plot(wave1[idx], spec1[idx], label='Salpeter')
plt.plot(wave2[idx], spec2[idx], label='Kroupa')
plt.plot(wave3[idx], spec3[idx], label='van Dokkum')
plt.legend(loc='best')
plt.xlabel('wavelength')
plt.ylabel('flux')
plt.title('spectrum')
```
